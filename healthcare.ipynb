```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv("D:/healthcare_dataset.csv")  # Update path_to_your_file.csv with the actual path
```

```python
# Explore the dataset
print(data.head())  # Check the first few rows of the dataset
print(data.info())  # Check the data types and missing values
```

```python
# a. Create a plot to find out if there is any correlation between Blood Group Type, Gender, and Medical Condition
plt.figure(figsize=(10, 6))
sns.countplot(x='Blood Group Type', hue='Medical Condition', data=data, palette='viridis')
plt.title('Blood Group Type vs Medical Condition')
plt.xlabel('Blood Group Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', hue='Medical Condition', data=data, palette='viridis')
plt.title('Gender vs Medical Condition')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# b. Find out the Average Billing Amount required to treat each medical Condition
average_billing = data.groupby('Medical Condition')['BillingAmount'].mean().reset_index()
print(average_billing)
```
